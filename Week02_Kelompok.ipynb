```python
from google.colab import drive

drive.mount('/content/drive')

import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img_before = cv.imread('/content/drive/MyDrive/pcvk_images/lisa_ktm.jpg')
img = cv.cvtColor(img_before, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

2. Tutup tiap bagian tertentu dari KTM tersebut menggunakan fungsi yang telah Anda pelajari. Kreasikan terkait dengan warna dan ukuran dari bentuk bentuknya

```python
# Points for the heart shape
points = np.array([[850, 600], [900, 475], [1100, 475], [1200, 600], [850, 750], [475, 600], [560, 475], [760, 475]], np.int32)
points = points.reshape((-1, 1, 2))


# Draw the red heart
cv.fillPoly(img, [points], color=(255, 0, 0))
# Draw the blue circle
cv.circle(img, (425, 150), 75, (0, 0, 255), -1)
# Draw the green oval
cv.ellipse(img, (510, 900), (110, 50), 0, 0, 360, (0, 255, 0), -1)

# Display the image
plt.imshow(img)
plt.title(" Kelompok 7 bagian Malang, bulan tahun dan NIP Direktur (NIM karena tidak ada)")
plt.axis('on') # Hide axes
plt.show()
```
